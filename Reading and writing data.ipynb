```
from pyspark.sql.types import *
```

```
users_schema= StructType([StructField("circuitId",IntegerType()),
                          StructField("circuitRef",StringType()),
                          StructField("name",StringType()),
                          StructField("location",StringType()),
                          StructField("country",StringType()),
                          StructField("lat",DoubleType()),
                          StructField("lng",DoubleType()),
                          StructField("alt",IntegerType()),
                          StructField("url",StringType()),

])
```

```
df=spark.read.option("header",True).schema(users_schema).csv("dbfs:/FileStore/tables/odinschool/circuits.csv")
```

```
display(df)
```

```
1. New column with today's date
2. new column with file path
3. drop url 
```

```
Tranformations
```

```
df1=df.withColumnRenamed("circuitId","circuit_id")\
 .withColumnRenamed("circuitRef","circuit_ref")\
.withColumn("ingestindate",current_date())\
.withColumn("path",input_file_name())\
.drop("url")
```

```
df1.write.mode("overwrite").option("header",True).csv("/FileStore/tables/odinschool/output/csv")
```

```
df1.write.mode("overwrite").parquet("/FileStore/tables/odinschool/output/parquet/")
```
